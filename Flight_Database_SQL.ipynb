```
%sql

/* Question 1 : Find the distinct flight numbers of all flights from Seattle to Boston by Alaska Airlines Inc. on Mondays. Also notice that, in the database, the city names include the state. So Seattle appears as Seattle WA. */

/* Results : 3 Rows */

SELECT DISTINCT f.flight_num from Flights_data f 
JOIN Carriers_data c ON c.cid = f.carrier_id
JOIN Weekdays_data w ON w.did = f.day_of_week_id
WHERE c.name = 'Alaska Airlines Inc.'
AND origin_city = 'Seattle WA'
AND dest_city = 'Boston MA'
AND w.day_of_week = 'Monday';
```

```
%sql 

/* Question 2 : Find the day of the week with the longest average arrival delay. Return the name of the day and the average delay.*/

/* Results : 1 Row */

SELECT day_of_week, AVG(F.arrival_delay) Arrival_Delay
FROM Flights_data f
JOIN Weekdays_data w 
ON w.did = f.day_of_week_id
GROUP BY w.did, w.day_of_week
ORDER BY Arrival_Delay DESC
LIMIT 1;
```

```
%sql

/* Question 3 : Find the names of all airlines that ever flew more than 1000 flights in one day. Return only the names. Do not return any duplicates. */

/* Result : 11 Rows */

SELECT DISTINCT c.name
FROM Flights_data f
JOIN Months_data m 
ON m.mid = f.month_id
JOIN Carriers_data c 
ON c.cid = f.carrier_id
GROUP BY f.day_of_month, f.year, m.month, c.name
HAVING COUNT(*) > 1000;
```

```
%sql

/* Question 4 : Find all airlines that had more than 0.5 percent of their flights out of Seattle be canceled. Return the name of the airline and the percentage of canceled flight out of Seattle. Order the results by the percentage of canceled flights in ascending order. */

/* Result : 6 Rows */

SELECT c.name, AVG(f.canceled) Cancelled_Percentage
FROM Flights_data f
JOIN Carriers_data c 
ON c.cid = f.carrier_id
WHERE f.origin_city = 'Seattle WA'
GROUP BY c.cid, c.name
HAVING Cancelled_Percentage > 0.005
ORDER BY Cancelled_Percentage ASC;
```
